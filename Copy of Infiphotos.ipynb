<h1>Capturing and collecting data by webcam</h1>

```python
cap = cv2.VideoCapture(0)
```

```python
if not cap.isOpened():
   print("unable to take footage")
```

```python
a = 1
while a<b:
 ret,frame = cap.read()
 if not ret:
       print("unable to open camera")
       break # Exit the loop if the camera cannot provide a frame
 cv2.imwrite(f"image_{a:02d}.jpg",frame)
 a += 1
 cap.release()
```
